```python
import pandas as pd
```

```python
df = pd.read_csv("Desktop/data science/areaprice.csv")
```

```python
df
```

Spiliting the data into independent and dependent variable

```python
x= df[["area"]]
```

```python
x
```

```python
y= df[["price"]]
```

```python
y
```

data visualization 

```python
import matplotlib.pyplot as plt
```

```python
plt.title("area vs price")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(x,y,color="red",s=100,ec="yellow")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[5000]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x
```
